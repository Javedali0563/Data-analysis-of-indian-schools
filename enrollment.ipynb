```python
filt3 = enrollment['Year'] == '2015-16'
```

```python
enrollment.loc[filt3].head()
```

```python
def extract_data(data):
    new_data = []
    for d in data:
        new_data.append(d)
    return new_data
        
```

```python
enroll_13 = extract_data(enrollment.loc[filt1,'Primary_Boys'])
enroll_14 = extract_data(enrollment.loc[filt2,'Primary_Boys'])
enroll_15 = extract_data(enrollment.loc[filt3,'Primary_Boys'])
states = extract_data(enrollment.loc[filt3, 'State_UT'])
len(enroll_15)
```

```python
#plt.bar(states,enroll_13)
x_indexes = np.arange(len(states))
plt.figure(figsize = (10,15))
hight = .4
plt.barh(x_indexes+hight,enroll_14,height = hight,label='2014')
plt.barh(x_indexes,enroll_15,height = hight,label='2015')
plt.legend()
plt.yticks(ticks=x_indexes,labels = states)
plt.tight_layout()
plt.title("Enrollment Percentage of each state")
plt.xlabel("Enrollment Percentage")
plt.ylabel("states")
plt.show()
```
